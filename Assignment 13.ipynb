```python
Ques 1.) What advantages do Excel spreadsheets have over CSV spreadsheets ?

Solution :
    
Excel spreadsheets offer several advantages over CSV (Comma-Separated Values) spreadsheets. Here are some key advantages of 
Excel :

1.) Formatting and Styling : Excel allows us to apply formatting options such as bold, italics, font styles, colors, and cell 
    borders. We can also use conditional formatting to highlight specific values based on certain criteria. This formatting 
    capability makes it easier to present data in a visually appealing and organized manner.

2.) Formulas and Functions : Excel provides a wide range of built-in formulas and functions that allow us to perform complex 
    calculations and automate tasks. We can use functions for mathematical operations, statistical analysis, data manipulation,
    and much more. These functionalities are not available in a plain CSV file.

3.) Charts and Graphs : Excel offers a variety of chart types and graphing options to visualize data in a meaningful way. We 
    can create bar charts, line graphs, pie charts, scatter plots, and many other types of charts to better understand patterns,
    trends, and relationships within our data. This visual representation can make data analysis easier and more accessible.

4.) Multiple Sheets and Workbooks : Excel allows us to have multiple sheets within a single workbook, and we can link and 
    reference data between sheets. This feature is helpful when dealing with complex data sets or organizing data into different
    categories. Additionally, we can create multiple workbooks and use features like formulas and macros to consolidate and 
    analyze data from various sources.

5.) Data Validation : Excel provides data validation tools that allow us to define rules and restrictions for data entry. We
    can set up dropdown lists, apply data type restrictions, and create custom validation rules to ensure data accuracy and 
    consistency. This helps prevent errors and maintains data integrity.

6.) Collaboration and Sharing : Excel supports collaboration features, allowing multiple users to work on the same spreadsheet
    simultaneously. We can track changes made by different users, leave comments, and protect sensitive data with passwords or
    file-level permissions. Excel also provides options for sharing spreadsheets via email or cloud storage platforms.

7.) Macros and Automation : Excel supports macros, which are recorded or programmed actions that automate repetitive tasks. We
    can record a series of steps and replay them later to save time and effort. Macros can be used for data cleaning, 
    formatting, calculations, and other tasks, making Excel a powerful tool for automation.

While CSV files have their own advantages, such as simplicity and compatibility with various software applications, Excel 
provides a broader range of features and capabilities for data analysis, presentation, and automation.
```
